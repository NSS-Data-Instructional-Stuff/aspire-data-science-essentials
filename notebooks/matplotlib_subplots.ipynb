### Matplotlib subplots demo - working with axes and figures
https://matplotlib.org/gallery/subplots_axes_and_figures/subplots_demo.html

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

### create some data to plot

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

####  `plt.subplots()` returns a tuple (fig, ax) and we see the default fig size and the number of axes is 1

```python
plt.subplots()
```

#### We can specify the figsize as an argument to `plt.subplots()` and assign the figure and ax to variables `fig` and `ax` 
##### Remember `ax` refers to a plot within the plotting area (or `fig`) so we can use it in the way we would use `plt` for a simple plot

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y)
ax.set_title('Simple plot');
```

### Two subplots with a shared x-axis
#### This time we create an array of axes (plots) called `ax_array` and pass arguments to `plt.subplots()` to specify *2* rows of plots in our figure. We have also told matplotlib that the plots will share an x-axis.

```python
f, ax_array = plt.subplots(nrows = 2, sharex=True)
f.suptitle('Sharing X axis')
ax_array[0].plot(x, y)
ax_array[1].scatter(x, y);
```

### Two subplots with a shared y-axis 
#### This time we set our plots to be drawn in 1 row and 2 columns.  Notice we return named axes this time (`ax1` and `ax2`) as a tuple.

```python
f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, sharey=True)
f.suptitle('Sharing Y axis')
ax1.plot(x, y)
ax2.scatter(x, y);
```

### How would you change the example above to plot the scatterplot on the left and the line plot on the right?
#### Also, notice how matplotlib has retained the figsize specified in the first plot. Change it to be 8 x 6 now.

```python
f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, sharey=True, figsize = (8, 6))
f.suptitle('Sharing Y axis')

ax2.plot(x, y)
ax1.scatter(x, y);
```

### Three subplots sharing *both* the x and y axes

```python
f, axarr = plt.subplots(nrows = 3, sharex=True, sharey=True)
f.suptitle('Sharing both axes')
axarr[0].plot(x, y)
axarr[1].scatter(x, y)
axarr[2].scatter(x, 2 * y ** 2 - 1, color='red');

# Bring subplots close to each other.
#f.subplots_adjust(hspace=0)

# # Hide x labels and tick labels for all but bottom plot.
# for ax in axarr:
#     ax.label_outer();
```

#### Creating plots in a grid with shared row and column axes 

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
f.suptitle('Sharing x per column, y per row')
ax1.plot(x, y)
ax2.scatter(x, y)
ax3.scatter(x, 2 * y ** 2 - 1, color='r')
ax4.plot(x, 2 * y ** 2 - 1, color='r');
```

#### Creating new data to demo `ax.twinx()`

```python
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)
```

#### First we'll plot `data1` in red 

```python
fig, ax1 = plt.subplots()

color = 'red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color);
```

#### Next we plot it again and add the second plot from `data2` in blue

```python
fig, ax1 = plt.subplots()

color = 'red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color2 = 'blue'
ax2.set_ylabel('sin', color=color2)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color2)
ax2.tick_params(axis='y', labelcolor=color2)

fig.tight_layout()
plt.savefig('double_plot.png', dpi = 400);  # otherwise the right y-label is slightly clipped
```
